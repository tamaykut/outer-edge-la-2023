```python
from moviepy.video.io.VideoFileClip import VideoFileClip
from moviepy.video.VideoClip import ImageClip
from moviepy.video.compositing.CompositeVideoClip import CompositeVideoClip
from moviepy.video.fx.resize import resize
import os
import numpy as np
from PIL import Image

img_path = 'NFT3.png'
img = Image.open(img_path)

video_path = 'NFT3-ad.mp4'
video = VideoFileClip(video_path)

# Resize the video to be 50% smaller
scale_factor = 0.5
video = resize(video, width=round(video.w * scale_factor), height=round(video.h * scale_factor))

# Create the ImageClip for the full duration of the video
image_clip = ImageClip(np.array(img).astype(np.uint8), duration=video.duration)

# Create the output video
composite = CompositeVideoClip([image_clip, video.set_position(('center', 'center'))])
output_path = 'NFT2-output.mp4'
composite.write_videofile(output_path, codec='libx264', audio_codec='aac')
```
